```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image

img = np.array(Image.open('trikoupi6.png'))

image = np.copy(img)
r = np.zeros((1670,1040))
x = 2
y = 2

rows,cols = image.shape
r[x:rows+x,y:cols+y] = image

plt.imshow(image,cmap="gray")
plt.show()
for i in range(1,r.shape[0]-1):
    for j in range(1,r.shape[1]-1):
        r[i][j] = r[i-1][j-1]*(1/9)+r[i-1][j]*(1/9)+r[i-1][j+1]*(1/9)+r[i][j-1]*(1/9)+r[i][j+1]*(1/9)+r[i+1][j-1]*(1/9)+r[i+1][j+1]*(1/9)

plt.imshow(r,cmap="gray")
plt.show()
```
